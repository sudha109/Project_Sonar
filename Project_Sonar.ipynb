```python
#imports necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#convert .csv file into dataframe
df_sonar=pd.read_csv('sonar.csv',header=None)
```

```python
df_sonar
```

```python
sns.heatmap(df_sonar.isnull(),annot=True)
plt.show()
```

```python
#dividing independent and dependent variables
x = df_sonar.iloc[:,0:60]
y = df_sonar.iloc[:,-1]
```

```python
#splitting the data
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
#using split rate of 20%
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
x_train
```

```python
#Applying MultinomialNB model,DecisionTree model,SVC and KNeighbors
#as in the dependent variable is categorical
#numerical algorithms cannot be performed on catogorical data
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

mnb=MultinomialNB()
mnb.fit(x_train,y_train)
predmnb=mnb.predict(x_test)

print(accuracy_score(y_test,predmnb))
print(confusion_matrix(y_test,predmnb))
print(classification_report(y_test,predmnb))
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
dtc.score(x_train,y_train)
preddtc=dtc.predict(x_test)
print(accuracy_score(y_test,preddtc))
print(confusion_matrix(y_test,preddtc))
print(classification_report(y_test,preddtc))
```

```python
from sklearn.svm import SVC

svc=SVC(kernel='rbf')
svc.fit(x_train,y_train)
svc.score(x_train,y_train)
predsvc=svc.predict(x_test)
print(accuracy_score(y_test,predsvc))
print(confusion_matrix(y_test,predsvc))
print(classification_report(y_test,predsvc))
```

```python
from sklearn.neighbors import KNeighborsClassifier
dtc=KNeighborsClassifier(n_neighbors=5,algorithm='brute')
dtc.fit(x_train,y_train)
dtc.score(x_train,y_train)
preddtc=dtc.predict(x_test)
print(accuracy_score(y_test,preddtc))
print(confusion_matrix(y_test,preddtc))
print(classification_report(y_test,preddtc))
```

```python
# saving pickle file
import pickle
filename='pickleKN.pkl'
pickle.dump(dtc,open(filename,'wb'))
```
